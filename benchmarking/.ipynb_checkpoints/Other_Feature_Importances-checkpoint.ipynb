# Napoles' Feature importance for Clustering's SHAP values

## Centroid (prototype)-based feature importance

```python
from prototypes import *
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import sys
sys.path.append('./')
from global_measures import *


##############################################################################
# Generate sample data
np.random.seed(0)

batch_size = 45
centers = [[1, 1], [-1, -1], [1, -1]]
n_clusters = len(centers)
X, labels_true = make_blobs(n_samples=1200, centers=centers, cluster_std=0.3)


kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

kmedians = KMedians(n_clusters=3)
kmedians.fit(X)

fuzzy_kmeans = FuzzyKMeans(n_clusters=3, m=2)
fuzzy_kmeans.fit(X)

fig = plt.figure()
colors = ['#4EACC5', '#FF9C34', '#4E9A06']

objects = (kmeans, kmedians, fuzzy_kmeans)

for i, obj in enumerate(objects):
    ax = fig.add_subplot(1, len(objects), i + 1)
    for k, col in zip(range(obj.n_clusters), colors):
        my_members = obj.labels_ == k
        cluster_center = obj.cluster_centers_[k]
        ax.plot(X[my_members, 0], X[my_members, 1], 'w',
                markerfacecolor=col, marker='.')
        ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
                                        markeredgecolor='k', markersize=6)
    ax.set_title(obj.__class__.__name__)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot ground truth with feature 0 and feature 1
plt.figure(figsize=(4, 4))
for k in np.unique(labels_true):
    plt.scatter(
        X[labels_true == k, 0], X[labels_true == k, 1],
        label=f"Cluster {k}", s=20
    )
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
# plt.title("Ground Truth Clusters")
plt.legend()
plt.tight_layout()
plt.show()
```

# Montavon's LRP for K-means

```python
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import contingency_matrix
from scipy.optimize import linear_sum_assignment
from sklearn.preprocessing import MinMaxScaler
from neuralized_kmeans import *
# sklearn KMeans

X = MinMaxScaler().fit_transform(X)

m = KMeans(n_clusters=3, random_state=77)
y_pred = m.fit_predict(X)
C = contingency_matrix(labels_true, y_pred)
_, best_match = linear_sum_assignment(-C.T)
y_aligned = np.array([best_match[i] for i in y_pred])

# Initialize neuralized model
X_tensor = torch.from_numpy(X)
logits = margins_kmeans(X_tensor, m)
nm = NeuralizedKMeans(m)

# Check match
assert torch.allclose(logits, nm(X_tensor)), "Mismatch in logits"
# Explain
R = neon(nm, X_tensor, beta=1.0)
```

```python
R.numpy().mean(axis=0)
```

```python
from global_measures import *
fig, ax = plt.subplots(1, 1, figsize=(3, 3))
sns.barplot(R.numpy().mean(axis=0)/R.numpy().mean(axis=0).sum(), ax=ax, orient='h')
ax.set_title('LRP')

# sns.barplot(psi, ax=ax[2], orient='h')
```
